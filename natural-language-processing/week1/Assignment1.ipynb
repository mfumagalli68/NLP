```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
import re
import pandas as pd
from ast import literal_eval

REPLACE_BY_SPACE_RE = re.compile('[/(){}\[\]\|@,;]')
BAD_SYMBOLS_RE = re.compile('[^0-9a-z #+_]')

print(REPLACE_BY_SPACE_RE)
```

```python
def preprocessing(data):

    stopword = set(stopwords.words('english'))
    data['title'] = data['title'].apply(lambda x: [item for item in x.split() if item not in stopword])
    data['title'] = data['title'].apply(lambda x: [i.lower() for i in x])
    data['title'] = data['title'].apply(lambda x: [re.sub(REPLACE_BY_SPACE_RE,'',i) for i in x])
    data['title'] = data['title'].apply(lambda x: [re.sub(BAD_SYMBOLS_RE,'',i) for i in x])
    
    
    return (data)

def read_data(path_file,test=False):


    if test:
        data = pd.read_csv(path_file, sep='\t')
        return (data)

    else:

        data = pd.read_csv(path_file,sep='\t')
        data['tags'] = data['tags'].apply(literal_eval)

        return (data)
```

```python
train = read_data('data/train.tsv')
test = read_data('data/test.tsv',test=True)
#train.head()
s = preprocessing(train)
s
```
